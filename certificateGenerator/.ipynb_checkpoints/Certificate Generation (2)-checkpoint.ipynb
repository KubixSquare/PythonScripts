```python
from PIL import Image, ImageDraw, ImageFont
import pandas as pd
```

```python
font = ImageFont.truetype(
    'C:\\Windows\\Fonts\\MyriadPro-Regular.otf',
    size=130 # font size according to resolution
)
color = 'rgb(76, 76, 78)' # black color
```

```python
data = pd.read_excel('KXSQ2020_Participants.xlsx')
data
```

```python
# all values in pixels (taken from mspaint)
Ys = [1400] # y coordinate of the blanks
Xstart = [1111] # X coordinate of beginning of blanks
Xend = [3343] # X coordinate of end of blanks

file_names = []
for i in range(len(data)):
    name = data.loc[i]['Name'].title()
    image = Image.open('template.jpg') # path to template
    name = name.title().strip()

    print('{}/{} - {}'.format(i+1, len(data), name))
        
    draw = ImageDraw.Draw(image)
        
    # centering text
    MAX_W = Xend[0] - Xstart[0] 
    w, h = draw.textsize(name, font=font)
    draw.text((Xstart[0]+(MAX_W-w)//2, Ys[0]), name, fill=color, font=font)
    
    file_name = f"{name.replace(' ', '')}_{data.loc[i]['Username']}.pdf"
    image.save(file_name, 'pdf')
    file_names.append(file_name)
    
print('Done')
```

```python
image
```

```python
data_ = pd.concat([data, pd.Series(file_names, name='FileName')], axis=1)
data_.to_excel('KXSQ2020_Participants_certificates.xlsx')
```

```python
form_data = pd.read_excel('KubixSquare Lockdown Contest Participation (Responses).xlsx')
form_data.drop_duplicates('Email Address', inplace=True)
form_data
```

```python
form_data['CodeChef Username'] = form_data['CodeChef Username'].str.lower().str.strip()
data_['Username'] = data_['Username'].str.lower().str.strip()
```

```python
merged = pd.merge(data_.rename(columns={'Username': 'CodeChef Username'}), form_data, how='outer')
merged = merged.sort_values('CodeChef Username').reset_index()
merged.to_excel('FormData_CodeChefData_merged.xlsx', columns=['Full Name', 'Email Address', 'CodeChef Username', 'FileName'])
```
